# differential learning rates

```python
model = resnet18()
```

```python
torch.optim.SGD([
                {'params':list(model.parameters())[:3], 'lr': 0.0000000001}, #stem
                {'params': model.layer1.parameters(), 'lr': 1e-5},
                {'params': model.layer2.parameters(), 'lr': 1e-4},
                {'params': model.layer3.parameters(), 'lr': 1e-3},
                {'params': model.layer4.parameters(), 'lr': 1e-2},
                {'params': model.fc.parameters(), 'lr': 1e-1}
            ], lr=1e-2, momentum=0.9)
```

# hooks

```python
def HOOK(*args, **kwargs):
    for a in args:
        print('-' * 30)
        print(a)
```

```python
model = resnet18()
conv2 = list(model.layer1.modules())[2]
conv2.register_forward_pre_hook(HOOK)
```
